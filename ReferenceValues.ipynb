```python
import pandas as pd
import numpy as np
import os
from matplotlib import cm
from colour import Color
```

```python
results = pd.read_excel('C:/Users/jtogt2/Notebook/umcu-stage/final_means_1.xlsx')
```

```python
references = pd.read_excel('C:/Users/jtogt2/Notebook/umcu-stage/reference_values.xlsx', index_col=0)
```

```python
references = references.dropna()
```

```python
red = Color("red")
colors = list(red.range_to(Color("green"),13))
colors.reverse()
```

```python
useful_columns = [i for i in results.columns if i in references.index]
useless_columns = [i for i in results.columns if i not in references.index and i not in ('cluster_assignment', 'cluster size', 'gender', 'age', 'n_people')]
new_cols = ['cluster_assignment', 'cluster size', 'n_people', 'gender', 'age'] + useful_columns + useless_columns
```

```python
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    return_li = []
    if s.name in references.index:
        minimum = references.loc[s.name]['min']
        maximum = references.loc[s.name]['max']
        for i in s:
            if minimum <= np.round(i, 3) <= maximum:
                return_li.append(f'background-color: {colors[0]}')
            elif (minimum/2 <= np.round(i, 3) <= minimum) or (maximum <= np.round(i, 3) <= maximum*2):
                return_li.append(f'background-color: {colors[4]}')
            elif minimum/3 <= np.round(i, 3) <= minimum or maximum <= np.round(i, 3) <= maximum*3:
                return_li.append(f'background-color: {colors[5]}')
            elif minimum/4 <= np.round(i, 3) <= minimum or maximum <= np.round(i, 3) <= maximum*4:
                return_li.append(f'background-color: {colors[6]}')
            elif minimum/5 <= np.round(i, 3) <= minimum or maximum <= np.round(i, 3) <= maximum*5:
                return_li.append(f'background-color: {colors[7]}')
            elif minimum/6 <= np.round(i, 3) <= minimum or maximum <= np.round(i, 3) <= maximum*6:
                return_li.append(f'background-color: {colors[8]}')
            elif minimum/7 <= np.round(i, 3) <= minimum or maximum <= np.round(i, 3) <= maximum*7:
                return_li.append(f'background-color: {colors[9]}')
            elif minimum/8 <= np.round(i, 3) <= minimum or maximum <= np.round(i, 3) <= maximum*8:
                return_li.append(f'background-color: {colors[10]}')
            elif minimum/9 <= np.round(i, 3) <= minimum or maximum <= np.round(i, 3) <= maximum*9:
                return_li.append(f'background-color: {colors[11]}')
            else:
                return_li.append(f'background-color: {colors[12]}')

    else:
        return_li = ['' for _ in s]
    return return_li
```

```python
results_copy = results.copy()
```

```python
results_ordered = results_copy[new_cols]
results_ordered
```

```python
results_colored = results_ordered.style.apply(highlight_max)
```
